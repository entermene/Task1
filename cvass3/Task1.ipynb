```python
import numpy as np
import cv2

def hough_circle_transform(image, radius_range, radius_increment, circle_center_increment, angle_increment, accumulator_threshold):
    # Convert the image to grayscale if it's not already
    if len(image.shape) == 3:
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray_image = image
    
    # Apply Canny edge detection
    edges = cv2.Canny(gray_image, 50, 150)
    
    # Initialize accumulator
    height, width = edges.shape
    max_radius = max(radius_range)
    accumulator = np.zeros((height, width, max_radius+1), dtype=np.uint64)
    
    # Iterate over each edge pixel
    for y in range(height):
        for x in range(width):
            if edges[y, x] > 0:  # If it's an edge pixel
                # Iterate over radius range
                for r in range(radius_range[0], radius_range[1] + 1, radius_increment):
                    # Iterate over angle range
                    for theta in range(0, 360, angle_increment):
                        a = int(x - r * np.cos(np.radians(theta)))
                        b = int(y - r * np.sin(np.radians(theta)))
                        if 0 <= a < width and 0 <= b < height:
                            accumulator[b, a, r] += 1
    
    # Find circle candidates above threshold
    candidates = []
    for r in range(radius_range[0], radius_range[1] + 1, radius_increment):
        for y in range(height):
            for x in range(width):
                if accumulator[y, x, r] > accumulator_threshold:
                    candidates.append([x, y, r, accumulator[y, x, r]])
    
    # Sort candidates by number of votes
    candidates.sort(key=lambda x: x[3], reverse=True)
    
    return candidates

# Example usage:
# Load the image
image = cv2.imread('Task1/images/smarties.png')

# Define parameters
radius_range = (10, 50)
radius_increment = 1
circle_center_increment = 1
angle_increment = 1
accumulator_threshold = 100

# Call the Hough transform function
circle_candidates = hough_circle_transform(image, radius_range, radius_increment, circle_center_increment, angle_increment, accumulator_threshold)

# Print circle candidates and their votes
for candidate in circle_candidates:
    print("Circle candidate:", candidate[:3], "Votes:", candidate[3])
```

```python
def non_maximum_suppression(circle_candidates, threshold):
    # Sort candidates by number of votes
    circle_candidates.sort(key=lambda x: x[3], reverse=False)
    
    # Perform non-maximum suppression
    suppressed_candidates = []
    for i in range(len(circle_candidates)):
        keep_candidate = True
        for j in range(i+1, len(circle_candidates)):
            # Calculate absolute differences in parameters
            diff_a = abs(circle_candidates[i][0] - circle_candidates[j][0])
            diff_b = abs(circle_candidates[i][1] - circle_candidates[j][1])
            diff_r = abs(circle_candidates[i][2] - circle_candidates[j][2])
            # Check if any difference exceeds the threshold
            if diff_a < threshold and diff_b < threshold and diff_r < threshold:
                keep_candidate = False
                break
        if keep_candidate:
            suppressed_candidates.append(circle_candidates[i])
    
    return suppressed_candidates

# Example usage:
# Define suppression threshold
threshold = 33

# Call non-maximum suppression function
filtered_candidates = non_maximum_suppression(circle_candidates, threshold)

# Print filtered circle candidates and their votes
for candidate in filtered_candidates:
    print("Circle candidate:", candidate[:3], "Votes:", candidate[3])
```

```python
# Use the opencv method to get the real ground truth to compare with myself's
def detect_circles(image, dp=1, minDist=20, param1=50, param2=30, minRadius=0, maxRadius=0):
    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Hough Circle Transform
    circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, dp, minDist,
                               param1=param1, param2=param2,
                               minRadius=minRadius, maxRadius=maxRadius)

    detected_circles = []
    if circles is not None:
        circles = np.uint16(np.around(circles))
        for circle in circles[0, :]:
            detected_circles.append([circle[0], circle[1], circle[2]])

    return detected_circles

# Example usage:
image = cv2.imread('Task1/images/smarties.png')

ground_truth_circles=detect_circles(image)
print("Detected circles:")

for circle in ground_truth_circles:
    print(circle)

def evaluate_detection_performance(ground_truth_circles, detected_circles):
    # Calculate the number of true positives, false positives, and false negatives
    true_positives = 0
    false_positives = 0
    false_negatives = 0

    # Iterate through each ground truth circle
    for ground_truth_circle in ground_truth_circles:
        # Check if the ground truth circle is detected
        detected = False
        for detected_circle in detected_circles:
            # Calculate the distance between centers
            distance = np.sqrt(((ground_truth_circle[0] - detected_circle[0]) ** 2 +
                               (ground_truth_circle[1] - detected_circle[1]) ** 2).astype(np.float64))
            # Check if the detected circle is within a certain threshold of the ground truth circle
            # if distance <= ground_truth_circle[2]:
            if distance <= 5:
                true_positives += 1
                detected = True
                break
        if not detected:
            false_negatives += 1

    # Calculate false positives
    false_positives = len(detected_circles) - true_positives

    # Calculate accuracy, false positive rate, and false negative rate
    total_circles = len(ground_truth_circles)
    accuracy = true_positives / total_circles
    false_positive_rate = false_positives / total_circles
    false_negative_rate = false_negatives / total_circles

    return accuracy, false_positive_rate, false_negative_rate

# Example usage:
# Assuming 'ground_truth_circles' is a list of ground truth circles and 'detected_circles' is a list of detected circles
accuracy, false_positive_rate, false_negative_rate = evaluate_detection_performance(ground_truth_circles, filtered_candidates)
print("Accuracy:", accuracy)
print("False Positive Rate:", false_positive_rate)
print("False Negative Rate:", false_negative_rate)


# Draw circles on the image
def draw_circles(image, circles):
    output_image = np.copy(image)
    if circles is not None:
        for circle in circles:
            center = (circle[0], circle[1])
            radius = circle[2]
            # Draw the circle
            cv2.circle(output_image, center, radius, (0, 0,255), 2)
    return output_image

# Example usage:
image = cv2.imread('Task1/images/smarties.png')

# Detect circles
detected_circles = detect_circles(image)

# Draw circles on the original image
image_with_circles = draw_circles(image, detected_circles)

# Display the result
cv2.imshow('Circles Detected', image_with_circles)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
